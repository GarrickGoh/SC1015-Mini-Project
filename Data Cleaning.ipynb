```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import calendar
```

```python
df = pd.concat(map(pd.read_csv, ['carpark_1_2018.csv','carpark_2_2018.csv','carpark_3_2018.csv','carpark_4_2018.csv','carpark_5_2018.csv','carpark_6_2018.csv','carpark_7_2018.csv','carpark_8_2018.csv','carpark_9_2018.csv','carpark_10_2018.csv','carpark_11_2018.csv','carpark_12_2018.csv',
                                'carpark_1_2019.csv','carpark_2_2019.csv','carpark_3_2019.csv','carpark_4_2019.csv','carpark_5_2019.csv','carpark_6_2019.csv','carpark_7_2019.csv','carpark_8_2019.csv','carpark_9_2019.csv','carpark_10_2019.csv','carpark_11_2019.csv','carpark_12_2019.csv',
                                'carpark_1_2020.csv','carpark_2_2020.csv','carpark_3_2020.csv','carpark_4_2020.csv','carpark_5_2020.csv','carpark_6_2020.csv','carpark_7_2020.csv','carpark_8_2020.csv','carpark_9_2020.csv','carpark_10_2020.csv','carpark_11_2020.csv','carpark_12_2020.csv',
                                 'carpark_1_2021.csv','carpark_2_2021.csv','carpark_3_2021.csv','carpark_4_2021.csv','carpark_5_2021.csv','carpark_6_2021.csv','carpark_7_2021.csv','carpark_8_2021.csv','carpark_9_2021.csv','carpark_10_2021.csv','carpark_11_2021.csv','carpark_12_2021.csv']), ignore_index=True)
holidays_df = pd.read_csv('sg_holidays.csv')
carpark_info = pd.read_csv('carpark-info.csv')
```

```python
df['update_datetime']= pd.to_datetime(df['update_datetime'])
df = df.drop(columns=['timestamp'])
df.head()
```

```python
carpark_arr = df.carpark_number.unique()
carpark_indexMap = pd.DataFrame(carpark_arr, columns = ['carpark_number'])
carpark_indexMap.insert(0, 'carpark_index', range(0, len(carpark_indexMap)))
carpark_indexMap
```

```python
carpark_info = carpark_info[carpark_info['car_park_no'].isin(carpark_arr)]
carpark_info = carpark_info.reset_index(drop=True)

print(carpark_info.shape)
display(carpark_info.head(5))
```

```python
# Even out datetime to nearest hour
df['update_datetime'] = df['update_datetime'].apply(lambda x: x.replace(minute=0, second=0))
df

# Create new hour_delta based on hours passed from Start Date
df['hour_delta'] = (df['update_datetime']  - df['update_datetime'].min())  / np.timedelta64(1,'h')
```

```python
# Data Cleaning
df['day'] = df['update_datetime'].dt.weekday
df['hour'] = df['update_datetime'].dt.hour
df.head(5)
```

```python
df = pd.merge(left=df,right=carpark_indexMap, left_on='carpark_number', right_on='carpark_number')
# df = df.drop(columns=['carpark_index_x', 'carpark_index_y'])
df
```

```python
holidays_df['Observance']= pd.to_datetime(holidays_df['Observance'])
holidays_df
```

```python
df["holiday"] = df['update_datetime'].isin(holidays_df['Observance']).astype(int)
df.head(5)
```

```python
df.to_csv("carpark_availability_cleaned.csv", sep=',', encoding='utf-8')
carpark_info.to_csv("carpark_info_cleaned.csv", sep=',', encoding='utf-8')
```
