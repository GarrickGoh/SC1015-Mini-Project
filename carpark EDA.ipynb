```python
# Load csv file
# Looking at 10 specific Carparks in Jurong Area
data = pd.read_csv('carpark_availability_cleaned.csv')
```

```python
#data cleaning
df = pd.DataFrame(data)
# Convert update_datetime column to datetime object
df['update_datetime'] = pd.to_datetime(df['update_datetime'])
display(df.sort_values(by='update_datetime').head())
df = df.drop_duplicates()
print(df.shape)
```

```python
display(df.head(20))
```

```python
print(df)
```

```python
df.total_lots = df.total_lots.astype('int64')
df.lots_available = df.lots_available.astype('int64')
```

```python
#total lots for the selected 10 carparks
total_lots = pd.DataFrame(df.groupby(by='carpark_number').total_lots.agg(np.mean)).sort_values(['total_lots'])
display(total_lots.head(10))
# some carparks lots have increased over the years or lower hence, not a whole number
# or maybe renovation of carpark or lots cause a decrease in lots.

#avg avail lots in a week, go by day. Least and most lots available.
#try to plot graph distance between carparks.
```

```python
#average lots availibility over 2018-2021 year
avail_lots = pd.DataFrame(df.groupby(by='carpark_number').lots_available.agg(np.mean)).sort_values(['lots_available'])
display(avail_lots.head(10))
```

```python
df['used_space'] = np.round((1 - df['lots_available']/df['total_lots'])*100, 2)
# Average used space in the collected period
used_space = pd.DataFrame(df.groupby(by='carpark_number').used_space.agg(np.mean)).sort_values(['used_space'])
display(used_space.head(10))
```

```python
def returnMonthlyAverageSlots(carpark):
    tempDf = df.copy()
    tempDf = tempDf[tempDf["carpark_number"] == carpark]
    tempDf = tempDf.groupby(pd.PeriodIndex(tempDf['update_datetime'], freq="M"))['lots_available'].mean()
    tempDf = tempDf.to_frame()
    
    tempDf['month'] = pd.PeriodIndex(tempDf.index).month
    tempDf['year'] = pd.PeriodIndex(tempDf.index).year
    return tempDf
```

```python
fig, axis = plt.subplots(ncols=2,nrows=5)
fig.set_size_inches(15, 10)

for i in range(0,10):
    if (i < len(carpark_arr)):
        carpark1Df = returnMonthlyAverageSlots(carpark_arr[i])
        sb.lineplot(x='month',y='lots_available',hue='year',data=carpark1Df, ax=axis[i%5,i%2]).set(title=str(carpark_arr[i]))
        
fig.tight_layout()
```

```python
# J67, JM31,J70 have large fluctuations for 2018
# J70,J63,J67,JM31,J62M have large fluctuations. For J69 between 7-9 months have large fluctuations
# J83M also has large fluctuations for 8-10 months. Overall for 2019 many carparks have fluctuations.
# JM31,JM2, large fluctuation for 2020.
# 2021 seems pretty similar throughout.
```
